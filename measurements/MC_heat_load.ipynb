```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vheater = np.array([10,20,30,40,50,60,70,80,90,110]) #uW
T = np.array([36,49,55,60,65,68,71,74,76,80]) #mK
T_squared = T**2
coef = np.polyfit(T_squared,vheater,1)
poly1d_fn = np.poly1d(coef) 
T_squared_axis = np.linspace(0,T_squared[-1]*1.1,300)
vheater_fitted = poly1d_fn(T_squared_axis)
```

```python
plt.rcParams['figure.figsize'] = [12.0, 8.0]
plt.rcParams['figure.dpi'] = 150
```

```python
fig, ax = plt.subplots()
ax.plot(T_squared,vheater, 'ro', T_squared_axis, vheater_fitted, '--b')
plt.xlim(T_squared_axis[0]*1.1, T_squared_axis[-1]*1.1)
plt.ylim(vheater_fitted[0]*1.1, vheater_fitted[-1]*1.1)
plt.title('Triton 4, heat load on MC, still heater off - FB 22 Jun 2021', size = 16)
plt.xlabel('[T$_{\mathrm{MC}}$ (mK)]$^2$',size= 14)
plt.ylabel('MC heater [$\mu$W]', size = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid(linewidth=1)

ax.annotate("Heat load ~" + "{0:.0f}".format(-vheater_fitted[0]) +' $\mu$W', xy=(0, vheater_fitted[0]), xytext=(T_squared[1], vheater_fitted[0]),
            arrowprops=dict(facecolor='black', shrink=0.01, width = 1), fontsize = 18)

            
```
